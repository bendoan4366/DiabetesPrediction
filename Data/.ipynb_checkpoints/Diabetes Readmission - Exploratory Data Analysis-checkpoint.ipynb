```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

from pathlib import Path

plt.style.use('fivethirtyeight')
```

```python
#import data and label keys

diabetes_path = 'dataset_diabetes/diabetic_data.csv'
dbd = pd.read_csv(diabetes_path)

key_path = 'dataset_diabetes/IDs_mapping.csv'
key = pd.read_csv(key_path)
```




First, we check the shape, feature types, and look at the dataframe at a high level



```python
dbd.shape 
```

```python
dbd.dtypes 
```

```python
#inspect data in each columns
dbd.head(15)
```

It looks like the weight column has a lot of missing values, so we inspect further using .describe() and value_counts

```python
dbd.describe(include = 'all')
dbd.loc[:,'weight'].value_counts()
```

A heavy majority of the weight column values are missing, so we're going to drop it from the dataframe

```python
dbd.drop('weight', axis =1, inplace=True)
```

```python
dbd.head(5)
```

Now we inspect other columns for missing values or values with '?' value

```python
for column in dbd:
    print(dbd.loc[:, column].value_counts(dropna = False))
```

It looks like the payer_code and medical_specialty have over half of their values missing, so we will drop those features as well

```python
dbd.drop('payer_code', axis =1, inplace=True)
dbd.drop('medical_specialty', axis =1, inplace=True)
```

Because age is represented in ranges, we're going to need dummy variables for each age group

```python
age_dummies = pd.get_dummies(dbd, columns=['age'])
age_dummies

#drop redundant column
age_dummies.drop('age_[0-10)', axis =1 )
age_dummies.head(5)
```

It looks like some of the columns have values that are keys to represent discrete data, so we're going to inspect the 'key' file now

```python
#visualize the key file
key.head(40)
```

It looks like admission_type_id, discharge_disposition_id, and admission_source_id, have values that need to be mapped, so we will do that next

```python
admin_id = {
    1: 'Emergency',
    2: 'Urgent',
    3: 'Elective',
    4: 'Newborn',
    5: None,
    6: None,
    7: 'Trauma',
    8: None
} 
age_dummies.loc[:,'admission_type_id'] = age_dummies.loc[:,'admission_type_id'].map(admin_id)
```

```python
#check for null values

age_dummies.loc[:, 'admission_type_id'].value_counts(dropna = False)

#it looks like there are enough values to keep this feature
```

```python
age_dummies.head()
```

```python
age_dummies.loc[:,'discharge_disposition_id'] = age_dummies.loc[:,'discharge_disposition_id'].map(dd_id)
```

```python
age_dummies.head()
```

```python
#check for null values

age_dummies.loc[:, 'discharge_disposition_id'].value_counts(dropna = False)

#it looks like there are enough values to keep this feature
```

```python
#Finally, we're going to map admission_source_id 
as_id = {
    1: 'Physician Referral',
    2: 'Clinic Referral',
    3: 'HMO Referral',
    4: 'Transfer from a hospital',
    5: 'Transfer from a Skilled Nursing Facility (SNF)',
    6: 'Transfer from another health care facility',
    7: 'Emergency Room',
    8: 'Court/Law Enforcement',
    9: None,
    10: 'Transfer from critial access hospital',
    11: 'Normal Delivery',
    12: 'Premature Delivery',
    13: 'Sick Baby',
    14: 'Extramural Birth',
    15: None,
    17: None,
    18: 'Transfer From Another Home Health Agency',
    19: 'Readmission to Same Home Health Agency',
    20: None,
    21: None,
    22: 'Transfer from hospital inpt/same fac reslt in a sep claim',
    23: 'Born inside this hospital',
    24: 'Born outside this hospital',
    25: 'Transfer from Ambulatory Surgery Center',
    26: 'Transfer from Hospice'
}
```

```python
age_dummies.loc[:,'admission_source_id'] = age_dummies.loc[:,'admission_source_id'].map(as_id)
age_dummies.head(10)
```

```python
#check for missing values

age_dummies.loc[:, 'admission_source_id'].value_counts(dropna = False)

#it looks like there are enough values to keep this feature
```
